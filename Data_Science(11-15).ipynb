```python
import matplotlib.pyplot as plt

categories = ['A','B','C','D']
sizes = [456 ,342, 800 ,231]

#first the integer values
#second the labels eg labels = city etc.

# '%' indicates that the value should be formatted as a percentage.
# 1.1f specifies the precision of the percentage--------
# 1: The number should have one digit before the decimal point.
# .1f: The number should have one digit after the decimal point.

# startangle=10: This parameter rotates the start of the pie chart by 10 degrees. By default, the start angle is 0 degrees.

plt.pie(sizes , labels = categories , autopct='%1.1f%%' , startangle =10)
plt.title('Pie Chart')

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [3, 5, 4, 6]
error = [0.2, 0.3, 0.1, 0.4]                    #error to axis coordinates either x or y which you mention.

# format to show datapoints = circle('o')
# capsize means top and bottom line of error bars ,can be zero i.e. no caps 

plt.errorbar(x, y, yerr=error, fmt='o', capsize=5)          #can be xerr= error or both 


plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Error Bar Chart')

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [3, 5, 4, 6]
error = [0.2, 0.3, 0.1, 0.4]                

plt.errorbar(x, y, xerr=error, yerr=error, fmt='o', capsize=5)    

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Error Bar Chart')

plt.show()
```

```python

categories = ['Audi', 'BMW', 'POLO']
values1 = [10, 20, 15]
values2 = [5, 15, 10]

plt.bar(categories, values1, label='Value 1')
plt.bar(categories, values2, bottom=values1, label='Value 2')

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart')
plt.legend()            #used to differentiate between the 2 labels eg blue -value1 ; orange-value2

plt.show()
```

```python
import numpy as np

#[np.random.normal(0, std, 100) - generate 3 sets of 100 random numbers from a normal distribution , with mean 0  
#for std in range(1, 4)]  - these 3 sets have mean 0 but std 1,2,3, respectively 


data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=False)

plt.xlabel('Value')
plt.title('Box Plot')

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [3, 5, 4, 6]

plt.plot(x, y)

plt.xticks([1, 2, 3, 4], ['One', 'Two', 'Three', 'Four'])                 #changes the x-axis tick labels from numeric values (1, 2, 3, 4) to string labels ('One', 'Two', 'Three', 'Four').
plt.yticks([3, 4, 5, 6], ['Low', 'Mid-Low', 'Mid-High', 'High'])          #changes the y-axis tick labels from numeric values (3, 4, 5, 6) to string labels ('Low', 'Mid-Low', 'Mid-High', 'High').
                                                                          # mention only those data points on the axis given . not all on the axes
plt.xlim(0, 5)                                                            #This sets the x-axis limits from 0 to 5.
plt.ylim(0, 7)                                                            #This sets the y-axis limits from 0 to 7.


plt.xlabel('X-axis')                  # X-label
plt.ylabel('Y-axis')                  # Y-label
plt.title('Plot with Custom Ticks')  #title
plt.plot(x, y, color='blue')        # Create a plot with custom line color
plt.plot(x, y, linestyle='--')     # dashed line
plt.plot(x, y, color='red', linestyle='dashed', marker='o', markersize=8, label='Data')

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [3, 5, 4, 6]

plt.plot(x, y)

plt.xticks([1, 2, 3, 4], ['One', 'Two', 'Three', 'Four'])
plt.yticks([3, 4, 5, 6], ['Low', 'Mid-Low', 'Mid-High', 'High'])
plt.xlim(0, 5)
plt.ylim(0, 7)


plt.xlabel('X-axis')                  # X-label
plt.ylabel('Y-axis')                  # Y-label
plt.title('Plot with Custom Ticks')  #title
plt.plot(x, y, color='blue',label = 'bluedata')        # Create a plot with custom line color
plt.plot(x, y, linestyle='--', label = 'dashed data')     # dashed line
plt.plot(x, y, color='red', linestyle='dashed', marker='o', markersize=8, label='Data')

plt.legend()                                              #labels mentioned due to this 

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)               # generate a specific sequence of random numbes thatis same for every time you run the code . 
                                 # you can give ny random value to random.seed() that is all giving their respective specific sequences diff from one another.
x = np.random.rand(30)               #Generates an array of 30 random numbers between 0 and 1 for the x-coordinates
y = np.random.rand(30)               #Generates an array of 30 random numbers between 0 and 1 for the y-coordinates 
colors = np.random.rand(30)          #Generates an array of 30 random numbers between 0 and 1 to be used for coloring the points


# c=colors ----assigns colors to the points based on the colors array
# s=100 ----sets the size of the points to 100.
# cmap='viridis' ----specifies the colormap to use for the points (this is for visually appealing plots.)
#edgecolors='k' ----- sets the edge color of the points to black ('k').

plt.scatter(x, y, c=colors, s=100, cmap='viridis', edgecolors='k')

plt.show()
```

```python
# Explode
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Audi", "BMW", "POLO", "KIA"]
myexplode = [0.2, 0, 0, 0]                             #Defines how much each slice should be "exploded" or separated from the center. 
                                                        # In this case Audi seperate by 0.2 units from the centre.

plt.pie(y, labels = mylabels, explode = myexplode)          #explosion effect according to the 'myexplode' array.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Audi", "BMW", "KIA", "SUZUKI"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)
plt.legend(title = "Cars")                  #Adds a legend to the pie chart with the title "Cars".legend helps to identify the slices based on their labels.
plt.show()
```
